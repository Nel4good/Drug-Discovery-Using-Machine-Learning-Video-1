```python
# A list of drug names
drugs = ["aspirin", "ibuprofen", "paracetamol", "caffeine"]

# Loop through and print each one
for drug in drugs:
    print("Drug:", drug)

# Print the first drug
print("First drug:", drugs[0])
```
